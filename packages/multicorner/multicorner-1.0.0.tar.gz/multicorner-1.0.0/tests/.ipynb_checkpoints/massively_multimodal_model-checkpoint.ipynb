```python
import numpy as np
import matplotlib.pyplot as plt
import corner
```

```python
np.random.seed(1)

N, M, n_samples_per_mode, wide_range = 15, 5, 5000, 100

mu = np.random.uniform(-wide_range, wide_range, (N, M))
covariances = [np.dot(A := np.random.rand(M, M), A.T) for _ in range(N)]
all_samples = np.vstack([np.random.multivariate_normal(mu[i], covariances[i], n_samples_per_mode) for i in range(N)])

corner.corner(all_samples, labels=[f"Dim {i+1}" for i in range(M)], show_titles=True)
plt.show()
```

```python
np.savez('dataset2.npz',all_samples)
```
