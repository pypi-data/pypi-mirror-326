```python
import numpy as np
import matplotlib.pyplot as plt
import corner

# Parameters
np.random.seed(1)

N = 20  # Number of modes (Gaussian distributions)
M = 2 # Number of dimensions
n_samples_per_mode = 5000
wide_range = 100  # Range for uniform distribution to draw centers (mu)

# Generate random centers for each mode
mu = np.random.uniform(-wide_range, wide_range, size=(N, M))

# Generate random covariance matrices for each mode
def generate_random_covariance_matrix(dim):
    A = np.random.rand(dim, dim)
    return np.dot(A, A.T)  # Ensure the covariance matrix is positive semi-definite

covariances = [generate_random_covariance_matrix(M) for _ in range(N)]

# Generate samples for each mode
samples = []
for i in range(N):
    samples.append(np.random.multivariate_normal(mean=mu[i], cov=covariances[i], size=n_samples_per_mode))

# Combine all samples into a single dataset
all_samples = np.vstack(samples)

# Plot the corner plot
corner.corner(all_samples, labels=[f"Dim {i+1}" for i in range(M)], show_titles=True)
plt.show()
```

```python
np.save('massively_multimodal_samples_lowD.npy',all_samples)
```
