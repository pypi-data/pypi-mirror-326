```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sample_rate(f):
    if f < 250:
        return 200e3
    if f < 500:
        return 500e3
    return 1e6
```

```python
frequencies = np.logspace(2, 4, 201)
sample_rates = np.vectorize(sample_rate)(frequencies)
```

```python
points = 2 * np.round(sample_rates / frequencies / 2)  # even numbers
```

```python
fig, ax = plt.subplots()
ax.semilogx(frequencies, points)
ax.grid()
```

```python
print("sine_table = [  # freq, ksps, points")
for p, sr in zip(points, sample_rates, strict=True):
    f = sr / p
    print(f"    ({int(round(sr / p))}, {int(round(sr / 1000))}, {int(round(p))}),")
print("]")
```
